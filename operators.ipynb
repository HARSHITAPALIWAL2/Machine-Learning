```python
# task : write a program in python to calculate sides of a rectangle

a= int(input("get value of a ="))
b= int(input("value of b = "))
c= int(input("value of c = "))

s  = (a+b+c)/2

area = (s*(s-a)*(s-b)*(s-c))
n=abs(area)
area = cmath.sqrt(n)
print(s)
print(area)
```
